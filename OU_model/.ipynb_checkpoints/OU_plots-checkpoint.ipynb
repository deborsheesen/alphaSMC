```python
from __future__ import division
%matplotlib inline
import numpy as np,  matplotlib.pyplot as plt, sys, gc
from time import time
from pylab import plot, show, legend
sys.path.append('/home/postdoc/dsen/Desktop/alphaSMC')
from PF import *
sys.path.append('/home/postdoc/dsen/Desktop/alphaSMC/OU_model')
from OU import *
plt.rcParams['figure.figsize'] = (15.0, 3.0)
```

```python
logNC_bpf = np.zeros((100,9))
testfn_bpf = np.zeros((100,9))
logNC_alpha = np.zeros((100, 9, 5))
testfn_alpha = np.zeros((100, 9, 5))

for i in range(10) :
    logNC_b = np.load("/home/postdoc/dsen/Desktop/alphaSMC/data/data_OU/logNC_bpf"+str(i+1)+".npy")
    logNC_a = np.load("/home/postdoc/dsen/Desktop/alphaSMC/data/data_OU/logNC_alpha"+str(i+1)+".npy")
    testfn_b = np.load("/home/postdoc/dsen/Desktop/alphaSMC/data/data_OU/testfn_bpf"+str(i+1)+".npy")
    testfn_a = np.load("/home/postdoc/dsen/Desktop/alphaSMC/data/data_OU/testfn_alpha"+str(i+1)+".npy")
    logNC_bpf[i*10:(i+1)*10] = logNC_b
    testfn_bpf[i*10:(i+1)*10] = testfn_b
    logNC_alpha[i*10:(i+1)*10] = logNC_a
    testfn_alpha[i*10:(i+1)*10] = testfn_a
```

```python
N_list = np.arange(2,11)*1000
C_list = 100*np.arange(1,6)+1
```

```python
plt.rcParams['figure.figsize'] = (5, 3)
plt.plot(N_list*np.var(logNC_bpf,0), "o-", label="BPF")
for i in range(5) :
    plt.plot(N_list*np.var(logNC_alpha[:,:,i],0), "o-", label="aplpa, C = "+str(100*(i+1)))
plt.legend(ncol=6)
```

```python
for i in range(5) :
    plt.plot(N_list*(np.var(logNC_alpha[:,:,i],0)-np.var(logNC_bpf,0)), "o-", label="aplpa, C = "+str(100*(i+1)))
plt.legend(ncol=5)
```

```python
for i in range(5) :
    plt.plot(N_list,np.var(logNC_alpha[:,:,i],0)/np.var(logNC_bpf,0), "o-", label="aplpa, C = "+str(100*(i+1)))
plt.legend(ncol=5)
```
