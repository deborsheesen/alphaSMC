### AIRPF:

```python
def within_island_resample(particles, theta, potential, y, A) :
    N = np.shape(particles)[0]
    pots = potential(y, particles, theta)
    W_out = np.zeros(N)
    for i in range(N) :
        weights = A[i]*pots
        W_out[i] = np.sum(weights)
        particles[i] = particles[npr.choice(a=N, size=1, p=weights/W_out[i])]
    return particles, W_out
```

```python
def augmented_island_resampling(particles, theta, W_out, y, A) :
    N = np.shape(particles)[0]
    S = np.shape(A)[0]
    idx = np.arange(N).astype(int)
    V = copy.deepcopy(W_out)
    for s in range(S) :
        V_old = copy.deepcopy(V)
        resampled_idx = copy.deepcopy(idx)
        for i in range(N) :
            weights = A[s,i]*V_old
            V[i] = np.sum(weights)
            idx[i] = npr.choice(a=resampled_idx, size=1, p=weights/np.sum(weights))
    return particles[idx]
```

```python
def AIRPF(data, theta, potential, propagate, test_fn, N) :
    N = np.shape(A)[1]
    y = data['y']
    T = np.shape(y)[0]
    _, _, _, test_fn_est, _, _ = initialise(data, theta, propagate, N)
    particles = np.zeros((N,T+1,len(x_0)))
    particles[:,0] = data['x_0']
    for t in trange(T) : 
        particles[:,t], W_out = within_island_resample(particles[:,t], theta, potential, y[t], A[1])
        particles[:,t] = augmented_island_resampling(particles[:,t], theta, W_out, y[t], A)
        particles[:,t+1] = propagate(particles[:,t], theta)
        test_fn_est[t] = np.mean(test_fn(particles[:,t+1]))
    return test_fn_est, particles
```

```python
S = 4
A = npr.rand(S,N,N)
for s in range(S) :
    for i in range(N) :
        A[s,i] = A[s,i]/np.sum(A[s,i])
        
test_fn_est, particles = AIRPF(data, theta, potential_SV, propagate_SV, test_fn, N)
```
